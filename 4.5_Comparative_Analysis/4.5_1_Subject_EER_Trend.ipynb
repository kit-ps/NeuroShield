```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
subjects = np.array([8, 16, 32, 64, 128, 230]).reshape(-1, 1)
eer_mean = np.array([23.84, 21.83, 18.16, 15.21,12.07, 7.98])
eer_std = np.array([10.52, 11.94, 10.18, 9.21,8.70, 7])

# Linear regression model
model = LinearRegression()
model.fit(np.log2(subjects), eer_mean)

# Predict EER for up to 16,384 subjects
pred_subjects = np.logspace(np.log2(8), np.log2(1500), 100, base=2).reshape(-1, 1)
predicted_eer = model.predict(np.log2(pred_subjects))
```

```python
pred_subjects
```

```python
# Update plot with increased font sizes for readability on A4
plt.figure(figsize=(6.3, 4))

# Font sizes
font_size_labels = 12
font_size_ticks = 10
font_size_legend = 10

# Actual data points with error bars
plt.errorbar(subjects.flatten(), eer_mean, yerr=eer_std, fmt='o', label="Actual Data", capsize=5, color="#0072B2")

# Prediction line
plt.plot(pred_subjects.flatten(), predicted_eer, label="Prediction", linestyle="--", color="#D55E00", linewidth=2)

strat_predection_area = 256
# Shaded area after prediction starts
plt.axvspan(strat_predection_area, pred_subjects.flatten()[-1], color='gray', alpha=0.2, label="Prediction Zone")

# Prediction start line
plt.axvline(x=strat_predection_area, color='gray', linestyle=':', label="Prediction Start")

# Labels and scale
plt.xlabel("Number of Training Subjects", fontsize=font_size_labels)
plt.ylabel("EER (%)", fontsize=font_size_labels)
plt.xscale("log", base=2)

# Adjust x-ticks to ensure proper display of two-digit powers
x_ticks = [2**i for i in range(3, 11)]  # Powers of 2 from 2^3 to 2^14
x_labels = [f"$2^{{{i}}}$" for i in range(3, 11)]  # Format labels with proper superscript braces
plt.xticks(x_ticks, labels=x_labels, fontsize=font_size_ticks)
plt.yticks(fontsize=font_size_ticks)

# Title and legend
#plt.title("EER Prediction Based on Number of Training Subjects", fontsize=font_size_title)
plt.legend(fontsize=font_size_legend, loc='upper right')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.tight_layout()
plt.savefig('EER_Predection.pdf', format='pdf', dpi=300)
plt.show()
```
